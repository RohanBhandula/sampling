```python
import numpy as np
import pandas as pd
df = pd.read_csv('Creditcard_data.csv')
```

## accuracy

```python
from sklearn.metrics import accuracy_score
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
def RandomForest(X_train, y_train, X_test, y_test):
  clf = RandomForestClassifier(n_estimators = 2)
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  return accuracy_score(y_test,y_pred)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
def KNN(X_train, y_train, X_test, y_test):
  knn = KNeighborsClassifier(n_neighbors = 5).fit(X_train, y_train)
  knn_pred = knn.predict(X_test)
  return accuracy_score(y_test,knn_pred)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
def DecisionTree(X_train, y_train, X_test, y_test):
  tree = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
  tree_pred = tree.predict(X_test)
  return accuracy_score(y_test,tree_pred)
```

# Logistic

```python
from sklearn.linear_model import LogisticRegression
def Logistic(X_train, y_train, X_test, y_test):
  clf = LogisticRegression()
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  return accuracy_score(y_test,y_pred)
```

# SVC

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
def SVC(X_train, y_train, X_test, y_test):
  svm = SGDClassifier(loss='hinge')
  svm.fit(X_train, y_train)
  y_pred = svm.predict(X_test)
  return accuracy_score(y_test,y_pred)
```

# Normalisation of amt

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the 'Amount' column
scaler.fit(df[['Amount']])

# Transform the 'Amount' column
df['Amount'] = scaler.transform(df[['Amount']])

# Drop the first column (assuming it's the index)
df = df.iloc[:, 1:]
df
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

# stratified

# cluster

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# select the features and target variable
X = dfnew
y = dfnew['Class']

# perform k-means clustering on the features
kmeans = KMeans(n_clusters=5, random_state=42).fit(X)
cluster_labels = kmeans.labels_

# create a DataFrame with the cluster labels and the target variable
cluster_df = pd.DataFrame({'cluster': cluster_labels, 'target': y})

# select a random sample of clusters
sample_clusters = np.random.choice(cluster_df['cluster'].unique(), size=5, replace=False)

# retrieve the observations from the selected clusters
sample_df = dfnew.loc[cluster_df['cluster'].isin(sample_clusters)]

# print the resulting sample DataFrame
print(sample_df.head())
```

```python
clusterX = sample_df.drop('Class', axis=1) #X_train
clustery = sample_df['Class'] #y_train

clusterRandomForest = RandomForest(clusterX,clustery,X_test,y_test)
clusterDecisionTree = DecisionTree(clusterX,clustery,X_test,y_test)
clusterKNN = KNN(clusterX,clustery,X_test,y_test)
clusterlogistic = Logistic(clusterX,clustery,X_test,y_test)
clustersvc = SVC(clusterX,clustery,X_test,y_test)

final3 = [clusterRandomForest,clusterDecisionTree,clusterKNN,clusterlogistic,clustersvc]
final3
```

# systematic

```python

# set the sampling interval
k = 4

# compute the number of samples
n = len(dfnew)

# compute the number of groups
num_groups = int(np.ceil(n / k))

# generate a list of starting indices for each group
start_indices = [i * k for i in range(num_groups)]

# retrieve the sample using systematic sampling
sample_df = dfnew.iloc[start_indices]

# print the resulting sample DataFrame
print(sample_df)
```

```python
systematicX = sample_df.drop('Class', axis=1) #X_train
systematicy = sample_df['Class'] #y_train

systematicRandomForest = RandomForest(systematicX,systematicy,X_test,y_test)
systematicDecisionTree = DecisionTree(systematicX,systematicy,X_test,y_test)
systematicKNN = KNN(systematicX,systematicy,X_test,y_test)
systematiclogistic = Logistic(systematicX,systematicy,X_test,y_test)
systematicsvc = SVC(systematicX,systematicy,X_test,y_test)

final4 = [systematicRandomForest,systematicDecisionTree,systematicKNN,systematiclogistic,systematicsvc]
final4
```
